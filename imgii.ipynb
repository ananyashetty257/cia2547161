```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread('butterfly.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
resized = cv2.resize(image, (200,200))
#TRANSLATION
#Parameters : input image ,M:tranformation matrix .dsize: size of output image
M = np.float32([[1,0,50],[0,1,30]])
translated = cv2.warpAffine(image,M,(image.shape[1], image.shape[0])) # 1 is width 0 is height and 2 is dimension
```

```python
#ROTATION
#cv2.getRotationMatrix2D(center,angle,scale) angle-A possitive value means counter clockwise rotation 
(h,w)= image.shape[:2]
center = (w//2,h//2)
M= cv2.getRotationMatrix2D(center,-45,1.0)
rotated = cv2.warpAffine(image,M,(w,h))
plt.figure(figsize=(15,12)) 

scaled=cv2.resize(image,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)

cropped=image[100:400,200:500]

fipped=cv2.flip(image,0)

pts1=np.float32([[50,50],[200,50],[50,200]])
pts2=np.float32([[10,100],[200,50],[100,250]])
M=cv2.getAffineTransform(pts1,pts2)
affine=cv2.warpAffine(image,M,(w,h))

plts1=np.float32([[56,65],[368,52],[28,387],[389,390]])
plts2=np.float32([[0,0],[300,0],[0,300],[300,300]])
M=cv2.getPerspectiveTransform(plts1,plts2)
perspective=cv2.warpPerspective(image,M,(300,300))

plt.subplot(3,4,1)
plt.imshow(resized)
plt.title('Resized Image')
plt.subplot(3,4,2)
plt.imshow(translated)
plt.title('Translated Image')
plt.subplot(3,4,3)
plt.imshow(rotated)
plt.title('Rotated Image')
plt.subplot(3,4,4)
plt.imshow(scaled)
plt.title('Scaled Image')
plt.subplot(3,4,5)
plt.imshow(cropped)
plt.title('Cropped Image')
plt.subplot(3,4,6)
plt.imshow(fipped)
plt.title('Flipped Image')
plt.subplot(3,4,7)
plt.imshow(affine)
plt.title('Affine Transformation')
plt.subplot(3,4,8)
plt.imshow(perspective)
plt.title('Perspective Transformation')
plt.show()
plt.figure(figsize=(15,12))
plt.imshow(image)
plt.title('Original Image')
plt.show()
```
