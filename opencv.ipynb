```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#read image in grayscale
image = cv2.imread('images2.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
print ("image shape (rows, columns): ", image.shape)
print("Pixel value at (0,0): ", image[40, 40])
```

```python
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
image1 = cv2.imread('imgii.jpeg')
rgb_image = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
```

```python
print("Image Properties")
print("Color Image Shape(H, W, C):", rgb_image.shape)
print("Grayscale Image Shape(H, W):", gray_image.shape)
print("Image data type:", rgb_image.dtype)
print("Total number of pixels in Color Image:", rgb_image.size)
print("Pixel value Range",rgb_image.min(),"to", rgb_image.max())
print("Average Pixel Intensity:", np.mean(image1))
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(rgb_image)
plt.title('Color Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off')

plt.tight_layout()
plt.show()  
```

```python

img = cv2.imread('images2.jpg')   

bgr  = img
rgb  = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
hsv  = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
lab  = cv2.cvtColor(bgr, cv2.COLOR_BGR2LAB)
ycrcb = cv2.cvtColor(bgr, cv2.COLOR_BGR2YCrCb)
gray_bgr = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
bgra = cv2.cvtColor(bgr, cv2.COLOR_BGR2BGRA)

plt.figure(figsize=(14,10))

plt.subplot(3,3,1), plt.imshow(rgb), plt.title('RGB'), plt.axis('off')
plt.subplot(3,3,2), plt.imshow(gray, cmap='gray'), plt.title('Gray'), plt.axis('off')
plt.subplot(3,3,3), plt.imshow(hsv), plt.title('HSV'), plt.axis('off')
plt.subplot(3,3,4), plt.imshow(lab), plt.title('LAB'), plt.axis('off')
plt.subplot(3,3,5), plt.imshow(ycrcb), plt.title('YCrCb'), plt.axis('off')
plt.subplot(3,3,6), plt.imshow(gray_bgr), plt.title('Gray  BGR'), plt.axis('off')
plt.subplot(3,3,7), plt.imshow(bgra), plt.title('BGRA'), plt.axis('off')

plt.tight_layout()
plt.show()
```
